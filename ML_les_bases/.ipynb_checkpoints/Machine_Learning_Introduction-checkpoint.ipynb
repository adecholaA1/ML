## ML - Les basiques 

ML est une branche de l'IA qui consiste à donner à une machine la capacité d'apprendre plutôt que de la programmer.

L'idée du ML est d'apprendre à partir des données. Pour cela il existe 3 grandes méthodes d'apprentissage : **supervisée, non supervisé, et par renforcement**


### A. Apprentissage supervisé :

L'essence de l'apprentissage supervisé est de développé un modèle qui représente au mieux les données de notre jeux de données, pour ensuite l'utiliser afin de réaliser de nouvelles prédictions.

La machine étudie les exaemples qu'on lui montre, et grâce à un système d'auto-évaluation et d'auto-amélioration, elle parvient à apprendre à réaliser la tâche qu'on lui demande d'effectuer.



Selon la nature de cette tâche, on distingue deux types de problèmes : 

**Les régressions** et **Les classifications**


#### 1. Régression

Ce sont des modèles utilisés pour prédire une variable quantitative $y$ à partir de données d'entrée $X$.

Exemple : Le prix d'une maison, le temps de trajet en voiture, le nombre de visiteurs quotidiens sur un site internet, le chiffre d'affaires d'une entreprise pour l'année prochaine, etc..

Ces modèles sont créés de manière **supervisée**, c'est à dire que l'on montre à la machine des exemples de données $(X, y)$ afin qu'elle développe un modèle permettant de relier $X$ à $y$.

Vous souhaitez prédire le prix d'une maison en fonction de sa surface habitable. Pour se faire : 


- On montre des exemples de maisons à notre machine

- Dans l'apprentissage supervisé, ces exemples de Q\R sont présentés à la machine sous forme de jeu de données $(X,y)$ où, 

    - $X$ représente les variables d'entrée
    - $y$ la sortie.
    
- Grâce à ce jeu de données, la machine est capable d'apprendre un modèle permettant de prédire la valeur $y$ en fonction de $X$.

- Pour ce faire, elle effectue une auto-évaluation en recherchant le modèle qui lui offre les meilleures performances par rapport au jeu de données fourni.

- Un fois ce modèle développé, il est possible de s'enservie pour faire les futures prédictions.

     - Par exemple:
     si nous avons, une nouvelle maison dont la surface habitable est de donnée (105 m$^2$), notre machine utilise le modèle pour prédire le prix de cette nouvelle maison.


#### - Régression linéaire:

Le modèle de régression linéaire se décrit par :

$$ f(x) = wx + b,$$

où $w$ et $b$ sont des paramètres qui contrôlent l'allure de $f$.

**L'objectif** est de trouver les paramètres $w$ et $b$ qui permettent au modèle de s'ajuster au mieux au jeu de données $(X,y)$.

Ceci se fait grâce à un algorithme d'optimisation qui va chercher à **minimiser** les erreurs entre le modèle et les points de données.

On commence donc par la mesure de ces erreurs : c'est la **fonction coût**

**La fonction coût**

Il s'agit de mesurer l'erreur entre chaque instant de données et notre modèle puis de faire la moyenne de toutes ces erreurs.

- On commence par calculer la différence entre la prédiction du modèle pour chaque point ($i$) et la valeur attendue $y^{(i)}$. On appelle cette différence le résidu 
    - $$R^{(i)} = f(x^{(i)}) - y^{(i)}$$
    
- Pour éviter que deux résidus de signes opposés s'annulent, on élève chaque résidu au carré, ce qui permet de n'obtenir que des résidus positifs. On obtient pur chaque point $(i)$:

    - $$ D^{(i)} = (f(x^{(i)}) - y^{(i)})^2$$
    
- Pour finir, on calcule la somme, puis la moyenne de toutes ces erreurs. Ce qui donne **la perte** globale du modèle : 

$$L = \frac{1}{2m} \sum_{i=1}^m (y^{(i)} - f(x^{(i)}))^2,$$

où $m$ désigne le nombre total de points de notre jeu de données.

**Allure de la fonction coût**

Lorsqu'on fait varier les paramètres $w$ et $b$ de notre modèle, on constate que **la fontion coût prend la forme d'une parabole qui aura un minimum**. Pour trouver ce minimum, on utilise l'algorithme de **la descente de gradient**.

**Algorithme de la descente de gradient**

Il consiste à calculer le gradient de la fonction coût c-à-d, comment celle-ci évolue lorsque $w$ et $b$ varient légèrement, pour ensuite faire u pas dans la direction où la fonction coût diminue.


   - **Description de l'algorithme de la descente de gradient**
   
     - Il démarre avec une initialisation avec des paramètres $(w_0, b_0)$ choisis au hasard. Il en résulte un modèle initial
     
     - À partir du point $(w_0, b_0)$, on calcule le gradient c-à-d, **la direction de la pente**.
     
     - On effectue alors un pas dans la direction de la descente. Cela nous donne de nouvelles valeurs pour $(w, b)$, notés $(w_1, b_1)$.
     
     - Avec ces nouvelles valeurs, nous obtenons également un nouveau modèle, qui se rapproche davantage des données $(X,y)$
     
     - En suite, on calcule à nouveau le gradient au nouveau point $(w_1, b_1)$, ce qui nous permet de poursuivre notre descente.
     
     - En continuant ainsi, on converge vers le minimum de la fonction coût, ce qui nous donne le modèle final.
     
Du point de vue mathématique on fait cette opération à chaque étape :

$$w_{t+1} = w_t - \eta \frac{\partial L}{\partial w_t} $$
$$ b_{t+1} = b_t - \eta \frac{\partial L}{\partial b_t},$$

$\eta$ est la vitesse d'apprentissage.

**Cet algorithme est largement utilisé en Apprentissage profond (à voir dans la suite)**


Pour développer un modèle de régression linéaire avec Python, nous allons utiliser le package **scikit-learn**.

```python
# Les imports
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

Pour commencer, simulons des données avec la librairie Numpy. Nous avons ici 50 points, distribués en suivant une relation $y = 1.5x + 3$

**Données**

```python
# Fixons la graine aléatoire pour une reproductibilité
np.random.seed(0)

m = 50

# Tirage de m point aléatoire suivant la loi gaussienne 
# et remise sous la forme (m_lignes,1_colonne)
X = np.random.randn(m).reshape(-1,1)

# y = 1.5 X + 3 + bruit
y = 1.5*X + 3 + np.random.randn(m).reshape(-1,1)

plt.scatter(X, y)
```

Nous allons utiliser le modèle **LinearRegression** de Sckit-learn. Il n'utilise par **la descente de gradient** vue plus haut, mais une autre méthode d'optimisation, plus rapide.

**Modèle**

```python
model = LinearRegression()
```

Une fois le modèle créé, nous utilisons la méthode **fit()** pour entraîner le modèle sur les données $(X, y)$.

```python
model.fit(X, y)
```

Une fois le modèle entraîné, il suffit de l'utiliser pour faire des prédictions à l'aide de la méthode **predict()**

```python
# Nouvelles données pour les prédictions
new_data = np.linspace(-3, 3).reshape(-1,1)


# Predictions
predictions = model.predict(new_data)


# Affichage 
plt.scatter(X, y)
plt.plot(new_data, predictions, c='r')
```

#### 2. Classification

On souhaite apprendre à reconnaître un chien ou un chat en nous basant sur sa taille et son poids : 

- Pour se faire, on fournit à la machine un jeu de données $(X,y)$ afin qu'elle s'entraîne.

- En étudiant ces données, **la machine trouve le modèle qui sépare aux mieux les deux classes de points**, c-à-d, celui qui ne commet aucune erreur sur le jeu de données qui lui est présenté.

- Avec ce modèle, la machine peut désormais faire de nouvelles prédictions.

- Lorsqu'on montre un animal que la machine n'a jamais vu auparavant, d'après le modèle développé à partir des données de références, cet animal peut être classé par exemple du côté des chiens ou des chats.


Ces modèles sont utilisés pour prédire la classe ou la catégorie à laquelle appartient un objet. 

Par exemple : 

- Prédire si un email est un spam ou non (2 classes possibles)

- Détecter si une cellule est cancéreuse ou non (2 classes possibles)

- Prédire le style de musique préféré d'une personne (environ 10 classes possibles)

- Reconnaître une plante et donner son nom (des milliers de classes possibles)

Ces moèdles sont créés de manière **supervisée** c-à-d que l'on montre à la machine des exemples $(X,y)$ afin qu'elle développe un modèle permettant de relier $X$ à $y$, **cela permet d'apprendre à séparer les différentes classes du jeu de données**.

**Mise en situation**

Imaginons qu'on souhaite prédire si un email est un spam ou non, en vous basant sur : **le nombre de liens** et **le nombre de fautes d'orthographe** qu'il contient.

- Pour cela, on dispose d'un jeu de données d'emails correspondant à la tâche que nous cherchons à accomplir.

- La méthode employée ici pour faire cette prédiction est celle de **plus proches voisins (nearest neighbors)** - méthode pour les problèmes de classification.

Il existe d'autres méthodes comme : 

- La Régression Logistique

- Les Supports Vector Machines

- Les Arbres de décision

- Les Random Forest

- ...

**Modèles du Nearest Neighbors (NN)**

Cet algorithme nécessite de garder en mémoire tous les points du jeu de données pour calculer la distance entre ces ppoints et un nouveau point dot on souhaite faire une prédiction.

Par exemple : 

- Supposons qu'on a un point dont on veut connaître la classe.

- On calcule la distance entre ce point et tous les autres points du jeu de donnaes. Il existe plusieurs méthodes pour calculer la distance entre deux points, mais nien souvent, on utilise la **distance euclidienne**.

- En classant chaque distance de la plus petite à la plus grande, on identifie le point du jeu de données le plus proche. Dès lors, on attribue notre point à la même classe que son plus proche voisin.

- Nous pouvons généraliser cette approche en appliquant cette opérations à tous les points de notre espace. Ce qui permet de mettre en avant **la frontière de deécision**, c-à-d la forntière qui sépare les deux classes.

Cet algorithme est largement utilisé dans de nombreuses applications, y compris chez *Spotify, Netflix, Amazon*, et bien d'autres.


**Généralisation : Le modèle K-Nearest Neighbors**

- L'algorithme NN (la forme 1 voisin) n'est pas robuste. Lorsqu'on fait face à un jeu de données un peu plus complexe avec des valeurs abérrantes dans chaque classes c-à- dire des emails spam qui se retouve dans la classe des non spams et vice versa. 

- Dans ce cas, la frontière de décision obtenue en suivant l'approche du plus proche voisin donne un résultat chaotique, qui ne correspond plus du tout à la tendance globale auquelle nous attendions. On pourrait se retrouver lors de la classification d'un nouveau email avec un email non spam dans la classe des emails spams.

- Pour éviter ce problème, on utilise généralement les k-voisins plus proches, pour ensuite assimiler notre point à la classe majoritaire parmi ces k-voisins.

Exemple : k = 3

- Notre point compte parmi ces 3 voisins les plus proches.

- Supposons qu'on a 2 points de la classe A et 1 point de la classe B.

- Notre point est donc assimilé à la classe A car en considérant les 3 plus proches voisnins, on a beaucoup plus de points de la classe A qui sont proche de notre point que des points de la classe B.

- Plus le nombre voisins augmente, moins la forntière de décision est sensibles aux valeurs abérrantes.

- Cependant il faut rester vigilant à ne pas trop augmenter le nombre de voisins, au risque de trop lisser la frontière de décision.

- pour un jeu de données dans lequel les deux classes prennent une forme de croissant, avec une légère superposition de certains points :

    - k = 1 n'est pas idéale, car elle est trop sensible aux valeurs abérrantes
    
    - k trop élevé aura pour éffet de tracer une frontière de décision qui ne correspond plus du tout à l'allure géniale de notre jeu de données
    
    - Il convient donc de trouver un juste milieur.
    
**Classification multiple**

On a également des problèmes de classification multiple (classification multiclasse).

Exemple : On a exemple de jeu de données comprenant 4 classes.

- L'alogorithme des k-voisins les plus proches fonctionnene parfaitement dans ce genre de situation, et nous n'avons rien à modifier dans notre algorithme pour obtenir les résultats attendus.

**Implémentations Python**

Nous avons besoin de la librairie de **Scikit-learn** et de son module **neighbors**. Au sein de ce module, on retrouve le modèle **KNeighborsClassifier**.

- Pour gérer les données, nous allons également utiliser la fonction *make blobs* provenant de *scikit-learn*.

```python
# Imports 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs             # Pour les données 
from sklearn.neighbors import KNeighborsClassifier  # Pour le modèle
```

Commençons par simuler des données. Nous créons deux classes $y$, avec 50 points, et deux variables X. 

Nous générons également des points dont les classes sont inconnues.

```python
# Données d'éntraînement 
X, y  = make_blobs(n_samples = 50,
                  centers = 2,
                  n_features = 2,
                  random_state = 0,
                  cluster_std = 1)

# Données de test
X_new = np.random.uniform(low = X.min(), high = X.max(), size = (10, 2))


plt.scatter(X[:, 0], X[:, 1], c = y, cmap = "bwr")
plt.scatter(X_new[:, 0], X_new[:, 1], c = "white", edgecolor = 'k')
```

Pour entraîner le modèle de KNeighborsClassifier, nous allons prendre **k=3** puis utiliser **fit()**

```python
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X, y)
```

Ensiuite, pour effectuer des prédictions et classer le nouveaux points de notre neu de données, on utilise la méthode **predict()**. On arrive ainsi à classer les nouveaux points du jeu de données.

```python
predictions = model.predict(X_new)

plt.scatter(X[:, 0], X[:, 1], c = y, alpha = 0.4)
plt.scatter(X_new[:, 0], X_new[:, 1], c = predictions, edgecolor = 'k')
```

### B. Apprentissage non supervisé.

Cette méthode consiste à laisser la machine apprendre par elle-mêm certaines structures présentes dans les données, sans la contraindre à apprendre une simple relation entrée/sortie.

Exemple : Nous pouvons lui présenter les groupes d'animaux vus précédemment sans préciser qu'il s'agit des chats ou de chiens, et lui demander simplement de les regrouper selon leur ressemblance

- Dans ce cas, nous fournissons uniquement les attributs $X$, sans indiquer la sortie $y$ attendue.

- En examinant simplement les attributs $X$ (tels que le poids, la taille, l'apparence, etc ...), la machine remarquera d'elle-même que certains animaux  se ressemble par leurs caractéristiques.

- Ainsi, elle sera capable de les regrouper, sans pour autant savoir s'il, s'agit de chats ou de chiens.

Dans cette section, on va apprendre à déveluopper des modèles de **clustering**. Ce type de modèle est utilisé lorsque l'on siuhaite classer nos données de manière **non supervisée**.

#### 1. Le Clustering

**Il consiste à regrouper nos données en nous basant uniquement sur leur ressemblance, sans se préocuper d'une quelconque classe "correcte" ou "incorrecte".**

Exemple : 

- classer des images ou des objets selon leur ressemblance

- Segmenter une base de données de clients, selon leurs habitudes de consommation

- Regrouper des documents selon leur contenu.

- etc...


C'est donc une méthode intéressante lorsqu'on souhaite laisser à la machine le pouvoir de proposer sa propre solution, et ainsi découvrir une approche différente de la nôtre.

**Mise en situation**

- Supposons qu'on dispose dans notre jeu de données différents types d'étoiles avec différentes formes(on peut aussi prendre différents types de biens immobiliers, d'individus, etc...).

- La tâche consiste à regrouper ces formes selon leur ressemblance, afin de créer trois groupes distincts.

- Les caractéristiques qui nous permettront de trier ces objets sont la taille et le nombre de branches de chaque forme.

- Ainsi, en mesurant ces attributs pour chaque objet et en les plaçant sur un graphique où sur l'axe des abscisses on a le nombre de branches et sur l'axe des ordonnées la taille.

- est-il nécessaire de superviser l'apprentissage de la machine en désignant des classes $y$ pour que celle-ci puisse classer ces points ? La réponse est non.

**Princioe du clustering**

Le principe du clustering est d'analyserles différentes variables $X$ qui caractérisent nos données, afin de regrouper les points en clusters, sans pour autant connaître la nature de ces points.

De nombreux algorithmes permettent de réaliser du clustering :

- Le K-Means Clustering

- Le Clustering hierarchique

- DBSCAN

- OPTICS

- etc.

**Nous parlerons ici de l'algorithme de K-Means Clustering**, qui est de loin le plus populaire et simple à comprendre.

**Fonctionnement du modèle de K-Means Clustering**

- C'est un algorithme itératif, dans lequel on demande à notre machine de trouver un nombre K de cluster au sein de notre jeu de données.

- Pour cela, elle place au hasard K points dans l'espace, puis déplace ces points pour qu'ils deviennent les barycentres de nos clusters

Exemple : Pour K = 3

- Pour commencer on place trois points aléatoires au sein de nos données. Ces points sont appelés **centroïdes** et sont les futurs centres de masse de clusters que l'on veut former.

- Ensuite, on associe chaque point de notre jeu de données au centroïde dont il est le plus proche, en calculant la distance euclidienne.

- Ensuite, on déplace chaque centroïde au centre de son groupe, ce qui explique son nom **centroïdes**. 

- Pour ce faire, on calcule la moyenne des points du cluster, et cette moyenne devient la nouvelle position du centroïde, d'où le nom **K-Means**.

- Comme la position de chaque centroïde a changé, les centroïdes ont désormais des plus proches voisins différents.

- Ainsi, nous réitérons notre algorithme...

- Nous réasssignos chaque point à son centroïde le plus proche (notez bien que cela produit des résultats différents de ceux de la première itération!).

- ... Puis on déplace chaque centroïde au centre de son groupe.

- Et cet algorithme de répète ainsi jusqu'à ce que les centroïdes ne bougent plus. C'est alors que l'algorithme se termine.


**C'est ainsi que fonctionne l'algorithme du K-Means clustering. Il suffit de déclarer le nombre de cluster que l'on souhaite obtenir.**

**Comment déterminer le nombre de centroïdes**

- Dans des complexes comme par exemple, la segmentation d'une base de données clients, comprenant des centaines de variables $X$, tels que leurs habitudes de consommation, leur âge, leur localisation, leur genre, etc. 

- Il n'est pas possible de visualiser les données pour déterminer le nombre optimal de clusters.

**Heureusement**, il existe une méthode permettant de trouver le nombre optimal de clusters sans avoir à observer directement notre jeu de données. Cette méthode s'appelle **la méthode du Coude**.

**La méthode du Coude**

Elle consiste à tester différentes valeurs de K pour l'algorithme deu K-Means Clustering, en mesurant à chauqe fois la variance finale de chaque clustering, puis en traçant sur un graphique l'évolution de cette variance en fonction de K pour identifier une forme de coude dans le graphique.

**Explication**

- Une **variance** est l'écart des points par rapport à leur moyenne c-à-d la distance globale des points d'un cluster par rapport à leur centroïde.

- Pour K = 1, on a une grande distance entre les points et leur centroïde, ce qui signifie que **la variance est donc élevée**

- On enregistre cette valeur de variance sur le graphique sous la valeur K = 1.

- Pour K = 2 clusters, la distance moyenne entre les points et leur centroïde respectif diminue par rapport au cas précédent.

- Nous ajoutons cette nouvelle valeur à notre graphique sous la valeur K=2

- Pour K=3 la variance continue de diminuer

- On continue K = 4,

-...

- Lorsque pour un K donné, la variance n'est pas très différente de la précente, cela nous indique que nous commençons à diviser un cluster qui avait déjà une variance assez faible.

- Nous entrons donc dans un phénomène de sur-optimisation, ce que nous souhaiterons éviter.

- En continuant, on observera sur notre graphe l'apparition d'une forme de coude. **C'est au point où le coude apparaît que K semble être optimal**

- La technique du coude **elbow method** est très populaire en clustering pour découvrir comment segmenter un jeu de données complexe et riche.

**Implémentation en Python**

Nous allons utiliser **Scikit-learn** en chargeant le modèle **KMeans** depuis le module **cluster**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

Nous générons un nombre aléatoire de clusters, ce nombre nous sera inconnu, et nous devrons le découvrir à l'aide de la méthode du coude

```python
k = np.random.choice([2, 3, 4, 5, 6, 7, 8]) # Choix qualqu'on pour avoir le X

X, _ = make_blobs(n_samples=200,
                 centers=k,
                 n_features=2,
                 cluster_std=0.5)
```

```python
plt.scatter(X[:, 0], X[:, 1])
```

Entraînons maintenant le modèle de KMeans pour diiférentes de de K de 1 à 9

```python
k_range = range(1, 10)
k_inertias = []

for k in k_range:
    model = KMeans(n_clusters=k).fit(X)
    k_inertias.append(model.inertia_) # Récupère les variances
    
plt.plot(k_range, k_inertias, lw=3)
```

```python
k
```

Le graphique résultant de cette expérience semble montrer un coude pour la valeur K = 4 (parfois la position du coude peut être différent).

Nous allons créer un modèle KMeans avec un nombre de clusters égal à 4

```python
model = KMeans(n_clusters = 4)
model.fit(X)

predictions = model.predict(X)

plt.scatter(X[:,0], X[:,1], c=predictions)
```

#### 2. Détection d'anomalies

Les modèles de detections d'anomalies sont utilisés pour **identifeir de nanière non supervisée, tout point s'écartant trop des autres points d'un jeu d'entrainement**

**L'idée** repose sur le fait qu'une anomalie, par définition, un évènement rare (sinon, cela devient une "norme"). 

- Ces caractéristiques X sont donc très éloignées de la majorité des autres points.

**Par exemple : **

- Une bouteille d'eau avec une forme très différente des autres sur une chaîne de production

- Une transaction bancaire frauduleuse, qui ne resemble en rien aux transations habituelles de votes compte

- Un piéton qui marche dans une zone où personne ne marche habituellement


**Mise en situation**

- Prenons une chaîne de fabrication de bouteilles d'eau. La grande majorité des bouteuilles ont une forme très semblable les unes aux autres. 

- Sur un graphique montrant le diamètre de chaque bouteille en fonction de sa taille, où les bouteuilles sont modélisées par un point, les points correspondant aux bouteuilles ayant des formes semblables seront regroupés

- Un bouteiulle très différente des autre passe devant nos capteurs, peu être facilement repérable car sa position sur le graphique sera très éloigné des autres. Cela prouve que ce genre de problème peut être résolu de manière non supervisée, sans avoir besoin de fournir d'informations sur une quelconque classe $y$ "normale" / "anomalie".

- Il pourrait même être contre-intuitif d'aborder un problème de détection d'anomalies avec une approche supervisée, comme la méthode de classification. On obtiendra une frontière de décision qui séparait les données comme illustré ci-dessus.

- À la place, **nous chercheons à développer un modèle qui puisse déclarer comme anomalie tout point déviant trop de la norme**. 

- Ainsi, notre fontière de décision devrait plutôt ressembler à :

     - un domaine qui entoure nos données normales 
     - et à l'extérieur de ce domaine on retrouve le/les points anomalies.
     
   
**Exemple d'algorithmes** permettant de résoudre ce genre de problème en machine learning :

- Local Outlier

- Isolation Forest

- Auto-Encoder


**Local Outlier Factor**

L'algorithme de Local Outlier Factor (LOF) **consiste à calculer l'écart de densité locale d'un point par rapport à celle de ses voisins.**

- Il considère comme **anomalies les points ayant une densité nettement inférieure à celle de l'ensemble des autres points**.

**Explication**

- Prenons un point au hasard dans notre jeu de données.

- L'algorithme du LOF commence par trouver les K-voisins les plus proches de ce point, pour K donné.

- Ensuite on trace un rayon d'action basé sur la distance au voisin le plus éloigné. C'est ce qu'on appelle **distance-k**

- Puis, pour chacun des k-voisins du point, nous procédons de la même manière (on calcule aussi la distance-k de ses k-voisins).

- Pour finir, notre algorithme calcule le rapport entre la distance-k du point choisi au hasard et la moyenne des distances-k de ces plus proches k-voisins.

- **Le nombre obtenu est le facteur local d'anomalie (Local Outlier Factor)**.

- **Plus ce facteur est élevé, plus il est probable que notre point soit considéré comme une anomalie**

- Dans le cas où les distances-k sont à peu prêt équivalentes, et leur rapport est proche de 1, on peu considérer que ce point n'est probablement pas une anomalie.


- En effectuant le calcul pour tous les points du jeu de données, nous sommes capable de détecter ceux dont le LOF est plus élevé que la moyenne, et ainsi de les marquer comme des anomalies.

### C. Apprnetissage par renforcement

Dans cette discipline, **on ne fournit pas de données $(X, y)$ à notre machine, mais on la programme afin qu'elle puisse générer ses propres expériences et apprendre de ses erreurs.**

**L'idée centrale** est donc de créer un **agent**, libre d'entreprendre des actions dans un **environnement**. Ces actions modifient **l'état** (state) de l'agent, et ce changement d'état s'accompagne d'une **récompense** (reward).

- Pour l'agent, le but est de maximiser ses récompenses, ce qui le pousse à apprendre quelles actions effectuer pour obtenir le plus de recompenses.

Cette discipline est très utilisée pour les applications suivantes : 

- Robotique
- Véhicule autonomes (Voitures, drones)
- Trading
- Algorithmes de prise de décision

**Mise en situation**

- Supposons qu'on programme un petit robot qu'il apprenne à sortir d'un labyrinthe le plus rapidement possible.

- Pour rendre les choses plus compliquées, le labyrinthe est généré aléatoirement.

- Nous pourrions pour cela construite un algorithme de recherche du chemin le plus court basé sur la théorie des graphes, mais cela reviendrait à coder explicitement le comportement de la machine? Et c'esy tout l'inverse du machine learning.

- À la place, nous voulons programmer la machine pour qu'elle développe elle-même la stratégie lui permettant de quitter le labyrinthe.

- Nous allons donc utikiser une approche d'apprentissage par renforcement.

- Notre agent sera le ppetit robot

- et le labyrinthe l'environnement dans lequel il évolue.

- L'état (state) correspond à la position de l'agent dans le labyrenthe, et les actions à ses déplacements possibles (haut, bas, gauche, droite)

- Depuis cet état, l'agent peut choisir une action, par exemple "aller en haut".

- En effectuant cette action, l'agent change son état, ce qui s'accompagne d'une récompense. Ici, nous attribuons à la machine un score de -1 point

- L'objectif étant de trouver la sortie le plus vite possible, nous infligeons un point négatif chaque fois que la machine effectue un déplacement.

- Son but étant de maximiser son score final, elle cherchera donc le moyen le plus rapide de rejoindre la sortie, préférence marquer -10 points plutôt que -15.

- En résumé, **notre objectif est** d'apprendre à la machine, quelle action effectuer lorsqu'elle se situe dans un état $s$ et produit une action $a$ c-à-d nous cherchons à développer une fonction $f(s) = a$ qui prend en entrée l'état $s$ et produit une action $a$. Cette fonction est couramment nommé **la politique d'action**.

- En RL, il existe **plusieurs méthodes permettant d'apprendre à développer cette politique d'action**. Celle que nous allons voir ici est l'approche appelée **Q-learning**

**Le fonctionnement du Q-Learning**

Le **Q-Learning** consiste à apprendre quelle est la **valeur** d'une action $A$ dans un état $S$ donné.

Exemple (le but du Q-Leraning): 

- Je mets habituellement 30 minutes en voiture pour aller à mon travail. Aujourd'hui, cette route est encombrée par le trafic. J'aperçois cependant qu'une autre route, un peu plus longue, est également possible. Ainsi, deux choix s'offrent à moi : l'action de tourner à gauche ou l'action de tourner à droite.

- Pour choisir laquelle emprunter, je vais évaluer la valeur de chacune. À mon avis, quel sera le temps de trajet total si je tourne à gauche ? Et si je tourne à droite ?

**Le but du Q-Learning** est donc d'apprendre à prédire la valeur de mes actions, en tenant compte de la situation dans laquelle je me trouve.

- Plus formellement, il s'agit donc d'apprendre une fonction Q qui prend en entrée un état $S$ et une action $A$, et qui prédit **le score final** (le meilleur score partant d'un étant et prenant une action) que l'on peut espérer obtenir si tout se passe pour le mieux par la suite. 

**L'apprnetissage de la fonction Q**

- Pour apprendre à prédire la valeur de ses actions, la machine doit explorer son environnement en s'y déplaçant au hasard, afin de générer des données (S, A, R) (état, action , récompense).

- Ces données sont ensuite utilisées pour mettre à jour **un tableau Q(S, A)** qui informe la machine des récompenses qu'elle est censée obtenir à l'avenir en choisissant telle ou telle action dans l'état présent

- Au début de son apprentissage, ce tableau est rempli de valeurs aléatoires

- Une formule, connue sous le nom de **l'équation de Bellman**, permet d'exprimer le score total que l'on peut espérer obtenir à l'avenir.

- Ce score est égal à la recompense obtenue immédiatement dans l'état $S$, à laquelle on ajoute le même score pour l'état suivant $S'$, si l'on choisit la meilleure action possible dans ce nouvel état.

$$Q(S, A) = \mathbb{E}(R(S') + \gamma \max Q(S', A'))$$

- $Q(S, A)$ : Récompense totales espérées

- $\mathbb{E}$  : l'espérance mathématique

- $R(S')$ : Récompense immédiate obtenue en allant à létat $S'$

- $\gamma$ : dévaluation

- $\max Q(S', A')$ : Récompenses totales espérées dans le nouvel état $S'$, en prenant la meilleure action $A'$

- **Explication**

    Dans l'analogie avec la voiture : 

    - L'équation de Bellman permet d'estimer que la valeur de tourner à gauche donne la récompense immédiate d'éviter le trafic, en plus de la valeur que vous allez obtenir dans votre prochain état.

    - Si vous tombez sur des bouchons 1km plus loin, le fait de tourner à gauche n'était peut-être pas la meilleure option.

    - À chque fois que la machine réalise une action, elle peut mettre à jour **sa table Q**, car elle reçoit une récompense associée à son chnagement d'état.

D'après la formule de Bellman, cette récompense $R(S')$ peut être utilisée pour recalculer la valeur de $Q(S,A)$ en se basant sur sa valeur actuelle et cette nouvelle donnée : 

$$ Q(S, A) = (1- \eta)Q(S, A) + \eta( R(S') + \gamma \max Q(S', A'))$$

- $Q(S,A)$ (à gauche) : nouvelle valeur

- $Q(S, A)$ (à droite) : valeur actuelle

- $\eta$ : vitesse d'apprentissage

- $R(S') + \gamma \max Q(S', A')$ : Equation de Bellman.

**Exploration et exploitation**

- Dès que la table Q commence à donner de bons résultats, la machine peut tomber dans le piège de suivre une politique d'action limitée, ignorant au passage les autres possibilités encore inexploités.

- Pour éviter ce phénomène, on définit dans l'algorithme du **Q-Learning** une probabilité $\varepsilon$ qui pousse la machine à continuer **d'exploiter** l'environnement des possiles, plutôt que de simpliment **exploiter** sa polituqe d'action déjà apprise.

- Ceci est le dernier éléùent qui compose le **Q-Learning**

```python
import gymnasium as gym
import numpy as np
```

Crée l’environnement (`env`) LunarLander-v2, un jeu où une fusée doit atterrir sur une zone cible.

**render_mode="human"** ouvre une fenêtre interactive (utile en local).

```python
env = gym.make("LunarLander-v3", render_mode="human")
```

`obs` (observation) : c’est l’état actuel de l’environnement.
Dans LunarLander, c’est un vecteur de 8 valeurs numériques qui décrivent la position, vitesse, angle, etc. du vaisseau.

Exemple : obs = [0.1, 0.5, -0.02, -0.1, 0.01, 0.0, 1.0, 0.0] : Ces valeurs sont ce que ton agent (ou ton IA) va utiliser pour prendre une décision.


`info` : ce sont des informations supplémentaires (métadonnées, pas toujours utiles pour l’IA)

`env.action_space` :  décrit toutes les actions possibles.

- Dans LunarLander-v3, il y a **4 actions possibles** :

    - 0 : ne rien faire
    - 1 : allumer le moteur gauche, afin de s'incliner à droite
    - 2 : allumer le moteur de bas, afin de ralentir sa pdescente
    - 3 : allumer le moteur droit afin de s'incliner à gauche
    
- Donc `env.action_space.sample()` : choisit une action au hasard.

Plus tard, l'agent utilisera un modèle de Machine Learning au lieu d’un choix aléatoire.

- Son **état** est décrit par sa position dans l'espace, ainsi que sa vitesse, son angle, et son contact avec le sol.

- Le système de **récompenses** est défini par plusieurs règles, parmi lesquelles 

    - Le vaisseau gagne 100 points s'il réussit à se poser sans dégâts
    
    - Le vaisseau perd 100 points s'il se crash
    
    - Le vaisseau perd des points chaque fois qu'il allume un moteur

`obs, reward, terminated, truncated, info = env.step(action)` : 

- Cette ligne exécute une étape de jeu :
    - `obs` : la nouvelle observation (nouvel état après avoir joué l’action).
    - `reward` : la récompense numérique (un score, positif ou négatif).
    
    Exemple : +100 si tu atterris bien, -100 si tu t’écrases.
    
    - `terminated` : vaut True si l’épisode se termine parce que l’objectif est atteint ou le vaisseau s’écrase.
    
    - `truncated` : vaut True si l’épisode s’arrête parce qu’on a atteint une limite de temps.
    - `info` : infos supplémentaires (par ex. diagnostics).
    
    
- `done` = terminated or truncated
Cela veut dire : l’épisode est terminé si le jeu est fini ou si le temps est écoulé.

```python
observation, info = env.reset()
done = False

for _ in range(1000):
    action = env.action_space.sample()   # action aléatoire
    observation, reward, terminated, truncated, info = env.step(action)
    
    if terminated or truncated:
        observation, info = env.reset()
env.close()
```

Nous commençons par initialiser notre environnement lunaire, ainsi que les différents hyper-paramètre de notre modèle : 

- learning rate
- le facteur de dévalorisation
- la valeur epsilon
- le nombre d'épisodes que la machine va jouer durant son apprentissage

```python
env = gym.make("LunarLander-v3", render_mode = "human")

# Hyper-paramètres
alpha = 0.1
gamma = 0.99
epsilon = 1.0
epsilon_decay = 0.995
epsilon_min = 0.01
episodes = 1000
max_steps = 1000
```

- Nous considérons que le vaisseau se déplace sur une grille de coordonnées.

- Nous allons découper l'espace des états du vaisseau spatial en plusieurs parties, afin de regrouper tous ces états dans la **table Q** (inialement remplie de 0).

```python
 state_bins = [np.linspace(-1.0, 1.0, 10) for _ in range(env.observation_space.shape[0])]
```

- `env.observation_space` : c’est l’espace des observations (l’état que tu reçois à chaque étape).

- `.shape[0]` = la dimension du vecteur d’état.

Exemple : pour LunarLander-v2, env.observation_space.shape[0] = 8 (car l’état est un vecteur de 8 nombres).

- `[state_bins = [...] for _ in range(env.observation_space.shape[0])]`: 

Ce code répète la même discrétisation (np.linspace(-1.0, 1.0, 10)) pour chaque dimension de l’état.

Si env.observation_space.shape[0] = 8, alors state_bins devient une liste de 8 arrays, chacun contenant 10 valeurs.

**But de ce code**

- Il sert à discrétiser un espace d’état continu.

Au lieu de manipuler des valeurs flottantes infinies (ex : positions, vitesses, angles), on les découpe en intervalles.

Exemple :

Si l’état est [0.2, -0.8] (juste 2 dimensions pour simplifier) :
Pour la première dimension, 0.2 tombe dans le “bin” (interval) autour de 0.111 (Entre 0.111 et 0.333.) → catégorie 5.
Pour la deuxième, -0.8 tombe dans le “bin” autour de -0.777 (est entre -1.0 et -0.777) → catégorie 1.
Donc, ton état continu [0.2, -0.8] est transformé en état discret (5, 1).

**En résumé** :

Ce code crée des bornes de découpage pour convertir un état continu (valeurs réelles) en un état discret (indices).
C’est souvent utilisé dans le Q-learning tabulaire, quand on ne peut pas stocker des Q-valeurs pour des états réels (infini).

```python
#np.prod(list(map(len, state_bins)))
```

```python
#len(state_bins[0])
```

```python
# Découpe 
n_bins = tuple(len(bins)+ 1 for bins in state_bins) # (11, 11, 11, 11, 11, 11, 11, 11)
#n_bins

q_table = np.zeros(n_bins + (env.action_space.n,))

```

Ici, q_table renvoie (10, 10, 10, 10, 4)

- Les 4 premières dimensions correspondent aux bins de l’état.
- La dernière dimension correspond aux actions possibles.


le `.n` correspond au nombre d’actions possibles quand l’espace d’action est discret.

Par exemple, pour l’état [3, 7, 1, 0] (après discretisation), q_table[3,7,1,0] est un tableau de 4 valeurs, une pour chaque action : [Q(s,a0), Q(s,a1), Q(s,a2), Q(s,a3)].

```python
def discretize_state(state):
    return tuple(np.digitize(state[i], state_bins[i]) for i in range(len(state)))
```

**Explications**

`state` : C’est un vecteur d’état continu, par exemple : [0.5, -0.1, 0.03, 0.8].

`state_bins` : C’est une liste de tableaux 1D qui définissent les limites de chaque bin pour chaque dimension de l’état.

Exemple pour une dimension :
state_bins[0] = [-1.0, -0.8, -0.6, ..., 0.8, 1.0]  # 10 bins

`np.digitize(state[i], state_bins[i])` :
Cette fonction prend la valeur continue state[i] et retourne l’indice du bin dans lequel elle tombe.

Si state[i] = 0.5 et tes bins sont [-1.0, -0.8, -0.6, ..., 0.8, 1.0], alors np.digitize renverra par exemple 8 si c’est le 8ᵉ bin.

`tuple(...)` : On convertit la liste d’indices de bins en tuple, car les indices seront utilisés pour indexer la Q-table.

- Nous allons maintenant créer la boucle pour entraîner notre vaisseau spatial à se poser san dégât.

Cette boucle comprendra 1000 épisodes, et chaque épisode comprend au maximum 1000 pas.

À chaque pas, le vaisseau spatial "lancer un dé", et en fonction du résultat, il choisira soir **d'explorer** son environnement en choisissant une action au hasard, soit **d'exploiter** sa politique d'action en choisissant l'action qui, selon lui, a la plus grande valeur (cette valeur est définie par la table-Q)

- Unfois son action effectuée, le vaisseau recevra un récompense, qu'il utilisera pour mettre à jour la table-Q en appliquant la mise à jour et l'équation de Bellman.

```python
for episode in range(episodes):
    state, _ = env.reset(seed=42)
    state = discretize_state(state)
    total_reward = 0
    
    for step in range(max_steps):
        # Epsilon-grredy policy
        if np.random.uniform(0, 1) < epsilon:
            action = env.action_space.sample() # Exploration
        else:
            action = np.argmax(q_table[state]) # Exploitation
        
        next_state, reward, terminated, truncated,  _ = env.step(action)
        next_state = discretize_state(next_state)
        
        old_value = q_table[state][action] # valeur Q actuelle
        next_max = np.max(q_table[next_state]) # valeur Q max de l'état S'
        
        # Mise à jour de la table q avec la formule de Bellman
        new_value = (1-alpha)*old_value + alpha*(reward + gamma*next_max)
        q_table[state][action] = new_value
        
        state = next_state
        total_reward +=reward
        
        if terminated or truncated:
            break
    # Reduction de epsilon
    if epsilon > epsilon_min:
        epsilon *= epsilon_decay
    
    print(f"Episode {episode + 1}, Total Reward: {total_reward}")
env.close()
```
